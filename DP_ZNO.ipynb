```python
list = ['Ukr', 'Hist', 'Math', 'Phys', 'Chem', 'Bio', 'Geo', 'Eng', 'Fra', 'Deu', 'Spa']
```

```python
for subject in list:
    print(pd.pivot_table(df, values=subject+'Ball', index=[subject+'TestStatus'] ,  aggfunc=[np.size, "count", min, max, np.median, np.mean]))
```

```python
dataframes = [0]*len(list)
for i, subject in enumerate(list):
    dataframes[i]=df[df[subject+'TestStatus'].notnull()]
```

```python
for i, subject in enumerate(list):
    dataframes[i] = dataframes[i][dataframes[i][subject+'TestStatus'] != 'Не з’явився']
    dataframes[i] = dataframes[i][dataframes[i][subject+'TestStatus'] != 'Не обрано 100-200']
    dataframes[i] = dataframes[i][dataframes[i][subject+'TestStatus'] != 'Анульовано']
```

```python
for i, subject in enumerate(list):
    print(pd.pivot_table(dataframes[i], values=subject+'Ball', index=[subject+'TestStatus'] ,  aggfunc=[np.size, "count", min, max, np.median, np.mean]))
```

```python
for i, subject in enumerate(list):
    dataframes[i]['SexTypeName'] = df['SexTypeName'].map({'чоловіча':1,'жіноча':0})
    dataframes[i]['TerTypeName'] = df['TerTypeName'].map({'місто':1,'селище, село':0, 'селище міського типу':0})
    dataframes[i][subject+'TestStatus'] = df[subject+'TestStatus'].map({'Зараховано':0, 'Не подолав поріг':1})
```

```python
for i, subject in enumerate(list):
    print(pd.pivot_table(dataframes[i], values=subject+'Ball', index=[subject+'TestStatus'] ,  aggfunc=[np.size, "count", min, max, np.median, np.mean]))
```

```python
for i, subject in enumerate(list):
    dataframes[i]['Age'] = 2021-dataframes[i].Birth
```

```python
for i, subject in enumerate(list):
    print(pd.pivot_table(dataframes[i], values='Birth', index=[subject+'TestStatus', 'SexTypeName'] ,  aggfunc=[np.size]))
```

```python
for i, subject in enumerate(list):
    print(pd.pivot_table(dataframes[i], values='Birth', index=[subject+'TestStatus', 'TerTypeName'] ,  aggfunc=[np.size]))
```

```python
def categorise(row):  
    if row['SexTypeName'] == 0 and row['TerTypeName'] == 0:
        return 0
    if row['SexTypeName'] == 1 and row['TerTypeName'] == 0:
        return 1
    if row['SexTypeName'] == 0  and row['TerTypeName'] == 1:
        return 2
    return 3
```

```python
for i, subject in enumerate(list):
    dataframes[i]['SexTer'] = dataframes[i].apply(lambda row: categorise(row), axis=1)
```

```python
result = []


print(result)
result += ['Name	Target	Protected	Number_of_records	distribution_in_groups	overall_prev	prev_(for_all_categories)	prev_diff'.split()]
for i, subject in enumerate(list):
    Name = 'ZNO_'+subject+'.csv'
    Target = subject+'TestStatus'
    Protected = ['SexTypeName', 'TerTypeName', 'SexTer']
    Number_of_records = len(dataframes[i])
    overall_prev = round(len(dataframes[i][dataframes[i][subject+'TestStatus'] == 1]) / len(dataframes[i]),3)

    for cat in Protected:
        distribution_in_groups = str(round((len(dataframes[i][dataframes[i][cat] == 1]))/Number_of_records *100))+'% '+':'+str(round((len(dataframes[i][dataframes[i][cat] == 0]))/Number_of_records *100)) +'% '
        
        prev_for_all_categories = str(round(len(dataframes[i][(dataframes[i][cat] == 1) &(dataframes[i][subject+'TestStatus'] == 1)]) / len(dataframes[i][dataframes[i][cat] == 1]),3))+':'+str(round(len(dataframes[i][(dataframes[i][cat] == 0) &(dataframes[i][subject+'TestStatus'] == 1)]) / len(dataframes[i][dataframes[i][cat] == 0]),3))
        
        prev_diff = abs(round(len(dataframes[i][(dataframes[i][cat] == 1) &(dataframes[i][subject+'TestStatus'] == 1)]) / len(dataframes[i][dataframes[i][cat] == 1])-len(dataframes[i][(dataframes[i][cat] == 0) &(dataframes[i][subject+'TestStatus'] == 1)]) / len(dataframes[i][dataframes[i][cat] == 0]),3))
        
        if cat != 'SexTer':
            result += [[Name , Target , cat, Number_of_records, distribution_in_groups , overall_prev , prev_for_all_categories, prev_diff ]]
        else:
            prev_diff_list =[0, 0, 0, 0]
            
            for j in range(4):
                prev_diff_list[j] = round(len(dataframes[i][(dataframes[i][cat] == j) &(dataframes[i][subject+'TestStatus'] == 1)]) / len(dataframes[i][dataframes[i][cat] == j]),3)
                
            distribution_in_groups = str(round((len(dataframes[i][dataframes[i][cat] == 3]))/Number_of_records *100))+'% '+':'+str(round((len(dataframes[i][dataframes[i][cat] == 2]))/Number_of_records *100)) +'% '+':'+str(round((len(dataframes[i][dataframes[i][cat] == 1]))/Number_of_records *100))+'% '+':'+str(round((len(dataframes[i][dataframes[i][cat] == 0]))/Number_of_records *100)) +'% '
        
            prev_for_all_categories = \
                str(prev_diff_list[3])+':'+\
                str(prev_diff_list[2])+':'+\
                str(prev_diff_list[1])+':'+\
                str(prev_diff_list[0])
            
            max_inter = max(prev_diff_list[3], prev_diff_list[2], prev_diff_list[1], prev_diff_list[0])
            min_inter = min(prev_diff_list[3], prev_diff_list[2], prev_diff_list[1], prev_diff_list[0])
            
            prev_diff = round(max_inter - min_inter,3)
            
            result += [[Name , Target , cat, Number_of_records, distribution_in_groups , overall_prev , prev_for_all_categories, prev_diff ]]
    
    
    # result += [[Name, Target, 'SexTer', len(dataframes[i]), str(len(dataframes[i][dataframes[i]['SexTer'] == 3]))+':'+str(len(dataframes[i][dataframes[i]['SexTer'] == 2]))+':'+str(len(dataframes[i][dataframes[i]['SexTer'] == 1]))+':'+str(len(dataframes[i][dataframes[i]['SexTer'] == 0]))], round(len(dataframes[i][dataframes[i][subject+'TestStatus'] == 1]) / len(dataframes[i]),3)]
print(result)
```
